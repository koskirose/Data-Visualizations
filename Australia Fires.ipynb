```python
# Import libraries
import numpy as np 
import pandas as pd 
import plotly as py
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
#Read the Data
df = pd.read_csv("fire.csv")
df.head(10)
```

```python
fig = px.density_mapbox(df, lat='latitude', lon='longitude', z='brightness', radius=10,
                        center=dict(lat=-25.2744, lon=133.7751), zoom=2,
                        mapbox_style="carto-positron",
                         
                       animation_frame='acq_date')
fig.update_layout(title_text ="Fires in Australia Aug 1 2019-Sept 30,2019 ",title_x=0.5,autosize=False, width=500, height=500)
fig.show()
```
